```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset (Salary dataset)
url = "https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.describe()
```

```python
# Rename columns for clarity
df.columns = ["YearsExperience", "Salary"]
```

```python
# Display first few rows
print("Dataset Preview:\n", df.head())
```

```python
# Step 1: Calculate Means
x_mean = df["YearsExperience"].mean()
y_mean = df["Salary"].mean()
print("\nMean of YearsExperience:", x_mean)
print("Mean of Salary:", y_mean)
```

```python
# Step 2: Calculate B1 (Slope)
numerator = ((df["YearsExperience"] - x_mean) * (df["Salary"] - y_mean)).sum()
denominator = ((df["YearsExperience"] - x_mean) ** 2).sum()
B1 = numerator / denominator
print("\nCalculated Slope (B1):", B1)
```

```python
# Step 3: Calculate B0 (Intercept)
B0 = y_mean - (B1 * x_mean)
print("Calculated Intercept (B0):", B0)
```

```python
# Step 4: Make Predictions
df["Predicted_Salary"] = B0 + B1 * df["YearsExperience"]
```

```python
# Step 5: Plot Regression Line
plt.scatter(df["YearsExperience"], df["Salary"], color='blue', label="Actual Data")
plt.plot(df["YearsExperience"], df["Predicted_Salary"], color='red', label="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```

```python
# Step 6: Calculate RMSE (Error)
rmse = np.sqrt(((df["Salary"] - df["Predicted_Salary"]) ** 2).mean())
print("\nRoot Mean Squared Error (RMSE):", rmse)
```
